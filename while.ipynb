```python
from bs4 import BeautifulSoup 
import requests
import lxml
```

```python
r= requests.get("https://www.worldometers.info/coronavirus/")
soup = BeautifulSoup(r.text, "lxml")
```

```python
print(soup.body.prettify())
```

```python
def parseNumber():
    r= requests.get("https://www.worldometers.info/coronavirus/")
    soup = BeautifulSoup(r.text, "lxml")
    number = soup.find_all('div', {'class': 'maincounter-number'})[0].find('span').text
    return number
while True:
    print('the current number is ' +str(parseNumber()))
```
